# Data collection
We aim to collect about 5000 headlines from the RSS feeds in `sources.csv`.

```python
from sourcing import from_urls, update_dataset
import pandas as pd

sources = pd.read_csv("sources.csv")
```

We collect the headlines and count them.

```python
feed = from_urls(sources.url)
feed.shape
```

Then we update the dataset and count the new total.

```python
update_dataset("data/raw_headlines.csv", feed)
pd.read_csv("data/raw_headlines.csv").shape
```
