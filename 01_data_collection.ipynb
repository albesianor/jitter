# Data collection
We aim to collect about 5000 headlines from the RSS feeds in `sources.csv`.

```python
from sourcing import from_urls, update_dataset
import pandas as pd

sources = pd.read_csv("sources.csv")
```

We collect the headlines and count them.

```python
feed = from_urls(sources.url)
feed.shape
```

Then we update the dataset and count the new total.

```python
update_dataset("data/raw_headlines.csv", feed)
pd.read_csv("data/raw_headlines.csv").shape
```

### Addressing the category unbalance while training the gatekeeper network
While this data is the one we care about, the labeled categories are largely unbalanced.  In fact, the RSS sources are already selected for relevance, so that the "irrelevant" category is already naturally small.  To avoid having the student network learn to select only relevant data, we will collect some extra data from irrelevant categories and add it to the training set for the gatekeeper network.

```python
irrelevant_sources = [
    "https://feeds.content.dowjones.io/public/rss/RSSLifestyle",
    "https://feeds.content.dowjones.io/public/rss/RSSArtsCulture",
    "https://feeds.content.dowjones.io/public/rss/RSSStyle",
    "https://feeds.content.dowjones.io/public/rss/rsssportsfeed",
    "https://rss.nytimes.com/services/xml/rss/nyt/Sports.xml",
    "https://rss.nytimes.com/services/xml/rss/nyt/Health.xml",
    "https://rss.nytimes.com/services/xml/rss/nyt/Arts.xml",
    "https://rss.nytimes.com/services/xml/rss/nyt/FashionandStyle.xml",
    "https://rss.nytimes.com/services/xml/rss/nyt/DiningandWine.xml",
    "https://rss.nytimes.com/services/xml/rss/nyt/Weddings.xml",
    "https://www.nytimes.com/services/xml/rss/nyt/Travel.xml",
]

irrelevant_feed = from_urls(irrelevant_sources)
irrelevant_feed.shape
```

```python
update_dataset("data/irrelevant_headlines.csv", feed)
pd.read_csv("data/irrelevant_headlines.csv").shape
```
